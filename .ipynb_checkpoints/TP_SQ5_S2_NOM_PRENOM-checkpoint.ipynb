### Modification de la taille de la carte

La taille des cartes affichés depuis le début de ce TP peut paraitre trop importante au vu de la taille de votre écran. Il est possible de modifier cette taille en redéfinissant les dimensions de la carte à afficher.

L'instruction qui permet de le faire dans la fonction folium.Map() est un couple de mots clés qui fait appel aux dimensions d'un rectangle: "width" et "height". Traduction: "longueur" et "largeur". ##  I.3. Placement d'un marqueur

Comme dans toute carte numérique que vous avez déjà utilisées, vous pouvez placer des marqueurs qui ont comme objectif de situer des lieux particuliers. Ces marqueurs fonctionnent comme des annotations. Vous pouvez leur attribuer une couleur, une forme et surtout une désignation précisant à l'utilisateur de votre carte le nom du lieu que vous indiquez, voire même une remarque particulière.  

La fonction de la librairie folium permettant de construire des marqueurs est la fonction **folium.Marker(xxxx)**.

Cette fonction place donc un marqueur sur la carte désignée au point de cordonnées (latitude,longitude) avec une couleur (paramètre color) choisie dans une palette de 147 nuances et un titre (paramètre popup).

Vous pouvez découvrir son utilisation dans le fragment de code suivant qui permet de placer un marqueur bleu au centre géographique de Paris contenant la désignation "Centre geographique de Paris". Lisez bien les commentaires pour comprendre chacune des lignes du code.

```python
carte_Paris = folium.Map(location=[48.85646, 2.34263], width='80%', height='80', tiles="OpenStreetMap", zoom_start=18)

folium.Marker([48.85646, 2.34263],          # Ajout d'un marqueur au point de cordonnées (latitude,longitude)
              popup="Centre geographique de Paris",   # Précision du contenu du marqueur
              icon=folium.Icon(color='blue')    # Atribution de la couleur
                                             ).add_to(carte_Paris)  # Ajout du marqueur dans la variable "carte_Paris", c'est à dire la carte voulue

carte_Paris
```

**Exécutez** ce code et **localisez** le marqueur que vous venez de construire sur votre nouvelle carte affichée.

**Cliquez** sur le marqueur en question. La désignation que vous lui avez attribuée dans le code apparait: "Centre géographique de Paris".

Pour rajouter d'autres marqueurs, vous devez appliquer la fonction folium.Marker(XXX) autant de fois que nécessaire dans votre programme.

Dans le fragment de code ci-dessous, on localise sur la même carte le centre géographique de Paris et le point zéro des routes de France.

```python
carte_Paris = folium.Map(location=[48.85646, 2.34263], width='80%', height='80', tiles="OpenStreetMap", zoom_start=15)

folium.Marker([48.85646, 2.34263], popup="Centre geographique de Paris", icon=folium.Icon(color='blue')).add_to(carte_Paris) 

folium.Marker([48.8533249, 2.3488596], popup="Point zero des routes de France ", icon=folium.Icon(color='red')).add_to(carte_Paris)  

carte_Paris
```

## I.4 Tracé d'une ligne brisée

Vous pouvez aussi tracer une ligne brisée qui relie des points de la carte définis par leurs coordonnées GPS.

La fonction de la librairie folium permettant de construire des lignes brisées est la fonction **folium.Polyline(xxxx)**.

Cette fonction relie donc des points sur la carte repérés par leurs cordonnées (latitude,longitude). On peut définir la couleur du tracé (paramètre color), son épaisseur (paramètre weight) et son intensité (paramètre opacity).  

Elle s'écrit sous cette forme: vous remarquerez que j'ai rassemblé les 2 points à relier dans une variable que j'ai nommé **points**

**points= [(48.85646, 2.34263), (48.8533249, 2.3488596)]**

**folium.PolyLine(points, color="blue", weight=2.5, opacity=0.8).add_to(carte_Paris)**


Vous pouvez découvrir son utilisation dans le fragment de code suivant qui permet de tracer une ligne brisée entre 3 lieux: le Centre geographique de Paris, le point zéro des routes de France et l'assemblée nationale qui se trouve sur la place de la concorde. 

Afin d'éviter de réécrire les coordonnées de ces points plusieurs fois dans le programme, j'ai pris la précaution d'utiliser des variables comme par exemple centre_geographique_Paris = (48.85646, 2.34263)

**Exécutez** le code et **visualisez** le tracé ainsi construit.

ATTENTION: ce tracé ne correspond pas du tout à un itinéraire. 

```python
# affectation des coordonnées GPS d'un lieu dans une variable
centre_geographique_Paris = (48.85646, 2.34263)
Point_zero = (48.8533249, 2.3488596)
Assemblee_nationale = (48.8617608, 2.3179738)

# Chargement d'une carte OSM
carte_Paris = folium.Map(location=centre_geographique_Paris, width='80%', height='80', tiles="OpenStreetMap", zoom_start=14)

# Placement des marqueurs
folium.Marker(centre_geographique_Paris, popup="Centre geographique de Paris", icon=folium.Icon(color='blue')).add_to(carte_Paris) 
folium.Marker(Point_zero, popup="Point zero des routes de France ", icon=folium.Icon(color='red')).add_to(carte_Paris)  
folium.Marker(Assemblee_nationale, popup="Point zero des routes de France ", icon=folium.Icon(color='orange')).add_to(carte_Paris)  

# Tracé d'une ligne brisée entre les 3 lieux prédéfinis
points = [centre_geographique_Paris, Point_zero, Assemblee_nationale, centre_geographique_Paris]
folium.PolyLine(points, color="blue", weight=2.5, opacity=0.8).add_to(carte_Paris)

# Affichage de la carte
carte_Paris
```

Vous remarquerez que la variable **points** fait apparaitre 2 fois **centre_géographique_Paris**: cela est nécessaire pour fermer la ligne brisée. Vous pouvez réexécuter ce programme en retirant une fois **centre_géographique_Paris** dans la variable **points** pour vous en convaincre.
